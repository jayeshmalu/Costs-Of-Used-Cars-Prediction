```python
import pandas as pd
```

### Reading Data

```python
train = pd.read_excel('Data_Train.xlsx')
train
```

### Data Cleaning

```python
train['New_Price'] = train['New_Price'].apply(lambda x: str(x).split(" ")[0])
train['New_Price'] = train.apply(lambda x: x['Price'] if x['New_Price']=='nan' else x['New_Price']  ,axis = 1)
train['New_Price'] = train['New_Price'].astype(float)
train
```

```python
train.isnull().sum()
```

```python
train = train[train['Power']!='null bhp']
train.dropna(inplace=True)
train.isnull().sum()
```

```python
train['Used Years'] = train['Year'].apply(lambda x: 2021-x)
train['Name1'] = train['Name'].apply(lambda x: str(x).split(" ")[0] + ' '+str(x).split(" ")[1])
train['Name2'] = train['Name'].apply(lambda x: str(x).split(" ")[0] + ' '+str(x).split(" ")[1]+' '+ str(x).split(" ")[2])
```

```python
print(train['Name'].value_counts())
print(train['Name1'].value_counts())
print(train['Name2'].value_counts())
```

```python
train['Power'] = train['Power'].apply(lambda x: str(x).split(" ")[0]).astype(float)
train['Engine'] = train['Engine'].apply(lambda x: str(x).split(" ")[0]).astype(float)
train['Mileage'] = train['Mileage'].apply(lambda x: str(x).split(" ")[0]).astype(float)
train.head()
```

```python
train.info()
```

```python
train1=train.copy()
```

### Label Encoding

```python
# Encoding categorical data
# Column Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train1['e_Name'] = le.fit_transform(train1.iloc[:,0])
train1['e_Name1'] = le.fit_transform(train1.iloc[:,14])
train1['e_Name2'] = le.fit_transform(train1.iloc[:,15])
train1['e_Location'] = le.fit_transform(train1.iloc[:,1])
train1['e_Transmission'] = le.fit_transform(train1.iloc[:,5])
train1
```

### Mapping Owner_Type

```python
owner_mapping = { "First":1,
             "Second":2,
             "Third":3,
            "Fourth & Above":4
                }
train1 = train1.replace({"Owner_Type":owner_mapping})
```

### One hot encoding

```python
### one hot encoding
train2 = pd.get_dummies(train1, columns = ['Fuel_Type'])
train2
```

### Dropping unnecessary columns

```python
train2.drop(['Year','Name','Name1','Name2','Location','Transmission'],axis=1,inplace = True)
```

```python
train2.info()
```

```python
train2.columns
```

### Preparing X & y

```python
y= train2['Price']
X = train2.drop(['Price'],axis=1)
X
```

### Feature Scaling

### Train-Test Split

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# from xgboost import XGBRegressor
# model = XGBRegressor(objective='reg:linear',n_estimators=10,seed=123)
# xg = model.fit(X_train,y_train)
# y_pred = xg.predict(X_test)
```

## Model Building

### Without hyperparameter tunning

```python
from sklearn.ensemble import ExtraTreesRegressor
et = ExtraTreesRegressor(n_estimators=100, random_state=0).fit(
   X_train, y_train)
y_pred = et.predict(X_test)
MAPE = (100-(abs(y_test-y_pred)*100/y_test)).mean()
print('Accuracy :',MAPE)
```

### With hyperparameter tunning

```python
from sklearn.ensemble import ExtraTreesRegressor
et = ExtraTreesRegressor(
    n_estimators=100,
    criterion='mse',
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    min_weight_fraction_leaf=0.0,
    max_features='auto',
    max_leaf_nodes=None,
    min_impurity_decrease=0.0,
    min_impurity_split=None,
    bootstrap=False,
    oob_score=False,
    n_jobs=-1,
    random_state=3595,
    verbose=0,
    warm_start=False,
    ccp_alpha=0.0,
    max_samples=None,
    ).fit(X_train, y_train)
y_pred = et.predict(X_test)
MAPE = (100-(abs(y_test-y_pred)*100/y_test)).mean()
print('Accuracy :',MAPE)
```
