```python
import pandas as pd
```

### Reading Data

```python
train = pd.read_excel('Data_Train.xlsx')
train
```

### Data Cleaning

```python
train['New_Price'] = train['New_Price'].apply(lambda x: str(x).split(" ")[0])
train['New_Price'] = train.apply(lambda x: x['Price'] if x['New_Price']=='nan' else x['New_Price']  ,axis = 1)
train['New_Price'] = train['New_Price'].astype(float)
train
```

```python
train.isnull().sum()
```

```python
train = train[train['Power']!='null bhp']
train.dropna(inplace=True)
train.isnull().sum()
```

```python
train['Used Years'] = train['Year'].apply(lambda x: 2021-x)
train['Name1'] = train['Name'].apply(lambda x: str(x).split(" ")[0] + ' '+str(x).split(" ")[1])
train['Name2'] = train['Name'].apply(lambda x: str(x).split(" ")[0] + ' '+str(x).split(" ")[1]+' '+ str(x).split(" ")[2])
```

```python
print(train['Name'].value_counts())
print(train['Name1'].value_counts())
print(train['Name2'].value_counts())
```

```python
train['Power'] = train['Power'].apply(lambda x: str(x).split(" ")[0]).astype(float)
train['Engine'] = train['Engine'].apply(lambda x: str(x).split(" ")[0]).astype(float)
train['Mileage'] = train['Mileage'].apply(lambda x: str(x).split(" ")[0]).astype(float)
train.head()
```

```python
train.info()
```

```python
train1=train.copy()
```

### Label Encoding

```python
# Encoding categorical data
# Column Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train1['e_Name'] = le.fit_transform(train1.iloc[:,0])
train1['e_Name1'] = le.fit_transform(train1.iloc[:,14])
train1['e_Name2'] = le.fit_transform(train1.iloc[:,15])
train1['e_Location'] = le.fit_transform(train1.iloc[:,1])
train1['e_Transmission'] = le.fit_transform(train1.iloc[:,5])
train1
```

### Mapping Owner_Type

```python
owner_mapping = { "First":1,
             "Second":2,
             "Third":3,
            "Fourth & Above":4
                }
train1 = train1.replace({"Owner_Type":owner_mapping})
```

### One hot encoding

```python
### one hot encoding
train2 = pd.get_dummies(train1, columns = ['Fuel_Type'])
train2
```

### Dropping unnecessary columns

```python
train2.drop(['Year','Name','Name1','Name2','Location','Transmission'],axis=1,inplace = True)
```
